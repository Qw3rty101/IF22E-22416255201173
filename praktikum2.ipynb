```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

distortions = []
K_range = range(1,10)
for k in K_range:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(df)
    distortions.append(kmeans.inertia_)

plt.plot(K_range, distortions, marker='o')
plt.title("Elbow Method for Optimal K (K-Means)")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Distortion")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k)
df['cluster'] = kmeans.fit_predict(df)

silhouette_avg = silhouette_score(df, df['cluster'])
print(f'Silhouette Score for K-Means {silhouette_avg}')
```

```python
import seaborn as sns

sns.scatterplot(x='sepal length (cm)',  y='sepal width (cm)', hue='cluster', data=df, palette='viridis', s=100)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=300, c='red', marker='x', label='Centroids')
plt.title('K-Means clustering of iris dataset')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.show()
```

```python
from sklearn_extra.cluster import KMedoids
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

distortions = []
K_range = range(1,10)
for k in K_range:
    kmedoid = KMedoids(n_clusters=k)
    kmedoid.fit(df)
    distortions.append(kmedoid.inertia_)

plt.plot(K_range, distortions, marker='o')
plt.title("Elbow Method for Optimal K (K-Medoids)")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Distortion")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

optimal_k = 3
kmedoid = KMedoids(n_clusters=optimal_k)
df['cluster'] = kmedoid.fit_predict(df)

silhouette_avg = silhouette_score(df, df['cluster'])
print(f'Silhouette Score for K-Medoid {silhouette_avg}')
```

```python
import seaborn as sns

sns.scatterplot(x='sepal length (cm)',  y='sepal width (cm)', hue='cluster', data=df, palette='viridis', s=100)
plt.scatter(kmedoid.cluster_centers_[:,0], kmedoid.cluster_centers_[:,1], s=300, c='red', marker='x', label='Centroids')
plt.title('K-Medoids clustering of iris dataset')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.show()
```
